#Data Normalisation

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

# Applying scaler() to all the columns except the 'yes-no' and 'dummy' variables
#num_vars = ['age','sex','cp','trestbps','chol','fbs','restecg','thalach','exang','oldpeak','slope','ca','thal']
num_vars = ['age','sex','cp','trestbps','chol','fbs','restecg','thalach','exang','slope','ca','thal']
heart_data[num_vars] = scaler.fit_transform(heart_data[num_vars])
df=pd.DataFrame(heart_data)

df.boxplot
```

#Data Split

```python
from sklearn.model_selection import train_test_split
df_train, df_test = train_test_split(heart_data, train_size = 0.7, test_size = 0.3, random_state = 100)
```

#Dividing the training data set into X and Y

```python
y_train = df_train.pop('target')
X_train = df_train
y_test = df_test.pop('target')
X_test = df_test
```

#Model Training

```python
from sklearn.metrics import accuracy_score
```

```python
from sklearn.linear_model import LogisticRegression
lr_model = LogisticRegression(max_iter=100)
lr_model.fit(X_train,y_train)
```

```python
y_pred_train = lr_model.predict(X_train)
y_pred_test = lr_model.predict(X_test)
```

```python
print(" Training Accuracy",lr_model.score(X_train, y_train))
print(" Testing Accuracy",lr_model.score(X_test, y_test))
```

```python
score_lrt = round(accuracy_score(y_pred_train,y_train)*100,2)
score_lr = round(accuracy_score(y_pred_test,y_test)*100,2)

print("The Training accuracy score achieved using Logistic Regression is: "+str(score_lrt)+" %")
print("The Testing accuracy score achieved using Logistic Regression is: "+str(score_lr)+" %")
```

Error Metrics Calculation

```python
import math
from sklearn.metrics import mean_squared_error 
from sklearn.metrics import mean_absolute_error


#Training Accuracies
rmse = math.sqrt(mean_squared_error(y_train, y_pred_train)) 
print('Root mean square error', rmse) 
mse = (mean_squared_error(y_train, y_pred_train)) 
print('Mean square error', mse) 
mae=mean_absolute_error(y_train, y_pred_train)
print('Mean absolute error', mae)
```

```python
#Testing Accuracies
rmse = math.sqrt(mean_squared_error(y_test, y_pred_test)) 
print('Root mean square error', rmse) 
mse = (mean_squared_error(y_test, y_pred_test)) 
print('Mean square error', mse) 
mae=mean_absolute_error(y_test, y_pred_test)
print('Mean absolute error', mae)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix_train=confusion_matrix(y_train, y_pred_train)
print(confusion_matrix_train)
sns.heatmap(confusion_matrix_train,annot=True)
```

```python
confusion_matrix_test=confusion_matrix(y_test, y_pred_test)
print(confusion_matrix_test) 
sns.heatmap(confusion_matrix_test,annot=True)
```

# Deploy Model

```python
list_of_columns=heart_data.columns
input_data=pd.DataFrame(columns=list_of_columns)
input_data.drop(['target'], axis='columns',inplace=True)

#age,sex,cp,trestbps,chol,fbs,restecg,thalach,exang,oldpeak,slope,ca,thal,target

input_data.at[0,'age']=int(input('Enter your age : '))
input_data.at[0,'sex']=int(input('Enter your Gender (0- female, 1 - male) : '))

input_data.at[0,'cp']=int(input(' Enter your chest pain type (0: typical angina 1: atypical angina 2: non-anginal pain 3: asymptomatic)'))
input_data.at[0,'trestbps']=int(input('Enter your resting blood pressure (in mm Hg on admission to the hospital) : '))

input_data.at[0,'chol']=int(input('Enter serum cholestoral in mg/dl: '))
input_data.at[0,'fbs']=int(input('Enter (fasting blood sugar > 120 mg/dl) (1 = true; 0 = false): '))

input_data.at[0,'restecg']=int(input('Enter resting electrocardiographic results Value 0: normal Value 1: having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV) Value 2: showing probable or definite left ventricular hypertrophy by Estes criteria: '))
input_data.at[0,'thalach']=int(input('Enter maximum heart rate achieved '))

input_data.at[0,'exang']=int(input('Enter exercise induced angina (1 = yes; 0 = no)  : '))
#input_data.at[0,'oldpeak']=float(input('Enter  ST depression induced by exercise relative to rest  : '))

input_data.at[0,'slope']=int(input('Enter the slope of the peak exercise ST segment Value 0: upsloping Value 1: flat Value 2: downsloping : '))
input_data.at[0,'ca']=int(input('Enter number of major vessels (0-3) colored by flourosopy : '))

input_data.at[0,'thal']=int(input('Enter thal: 1 = normal; 2 = fixed defect; 3 = reversable defect and the label : '))
```
